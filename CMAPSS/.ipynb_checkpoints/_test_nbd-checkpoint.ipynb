```python
import time
import os
import gzip
import pickle
import platform
import numpy as np
import pandas as pd

## sourcing data (require psycopg2)
import funforsql_blagdon as ffs

pklhp = pickle.HIGHEST_PROTOCOL
```

```python
# meta settings
workers = 4
#batch_size = 6*4096
#epochs = 1
timeunit = 'week' # valid choices: day, week, month, quarter
incl_alltrans_percust = False

# names and stuff
#purpose = 'gprun'
device = open('_devicename_.txt').readline()
#modelspec_fname = 'spec05.py'

# db name
skusfname = 'sodapopskus.txt'
dbid = skusfname
```

```python
transactions = ffs.gettransactions_cached(skusfname, 
                                          timeunit='week', 
                                          incl_alltrans_percust=incl_alltrans_percust,
                                          verbose=True)

if device == 'macbookpro':
    import matplotlib.pyplot as plt
    def get_date_counts(obj):
        orddate, subset = obj
        return (orddate, len(subset))    
    groupedby_orddate = transactions.groupby('orddate_index')
    transactions_perorddate = np.array(list(map(get_date_counts, list(groupedby_orddate))))
    plt.plot(transactions_perorddate[:,0], transactions_perorddate[:,1])
    plt.show()

discardfirst = 300
dates = transactions.loc[:, 'orddate_index']
dates = dates - discardfirst
transactions.loc[:, 'orddate_index'] = dates
transactions = transactions.loc[dates>=0,:]

com, split, fin = ffs.gettimes(transactions)
print ('com:%i, split:%i, fin:%i' % (com, split, fin))

transtrain = ffs.getdf_intimerange(transactions, com, split)
#print (transtrain.head(100))

uniqueskus = np.sort(np.unique(transtrain.loc[:, 'product_sku']))
```

### modify source db type

```python
#skudex = None
skudex = 0
useskus = uniqueskus
if skudex is not None:
    useskus = np.array([uniqueskus[skudex]])
    
print (useskus)
```

```python
useobs = np.in1d(transtrain.loc[:,'product_sku'], useskus)
if not str(useskus) == str(uniqueskus):
    for sku in useskus: 
        dbid += '_' + str(sku)
print ('dbid:', dbid)

transtrain = transtrain.loc[useobs,:]
print ('unique skus in transtrain:', len(np.unique(transtrain.loc[:,'product_sku'])))
```

```python
def getfrt(enddate, groupbyobj):
    cid, subset = groupbyobj
    purchdates = subset.loc[:, 'orddate_index'].as_matrix()
    frequency = max((0, len(subset) - 1)) 
    age = enddate - min(purchdates) # age of customer... today
    recency = max(purchdates) - min(purchdates) # recency of purchase
    if age < recency:
        print (cid, subset)
    return (cid, frequency, recency, age)

def worker(groupbyobj):
    return getfrt(split, groupbyobj)

from multiprocessing import Pool
with Pool(workers) as p:
    resm = np.array(list(p.map(worker, transtrain.groupby('custid'))))
print (resm.shape)
    
    
d = {'frequency': resm[:,1], 'recency': resm[:,2], 'age': resm[:,3]}
frt_training = pd.DataFrame(d, index=resm[:,0])
```

```python
problem = frt_training.loc[:,'recency'] > frt_training.loc[:, 'age']
print (frt_training.iloc[np.where(problem)[0],:])
```

```python
from lifetimes import BetaGeoFitter
bgf = BetaGeoFitter(penalizer_coef=0.0)
bgf.fit(frt_training['frequency'], 
        frt_training['recency'], 
        frt_training['age'])
print(bgf)
```

```python
pred = bgf.conditional_probability_alive(frt_training['frequency'], 
                                         frt_training['recency'], 
                                         frt_training['age'])
```

```python
transtest = ffs.getdf_intimerange(transactions, split, fin)
useobs = np.in1d(transtest.loc[:,'product_sku'], useskus)
cidtest = ffs.getcustid(transtest.loc[useobs,:])
```

```python
eval_cid    = np.array(frt_training.index.values)
eval_metric = pred.as_matrix()
eval_truth  = np.in1d(eval_cid, cidtest)
print (eval_cid.shape, eval_metric.shape, eval_truth.shape)

evaldf = pd.DataFrame({'cid'   : eval_cid,
                       'metric': eval_metric,
                       'truth' : eval_truth})

pickle.dump(evaldf, 
            gzip.open('nbd_DBID_%s_SKUDEX_%s_eval.pkl'%\
                      (str(dbid), str(skudex)) + str(pklhp), 'wb'),
            -1)
```
